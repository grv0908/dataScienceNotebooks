# MatplotLib

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Line Chart

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.162, 5.263, 6.972]
```

```python
plt.plot(year, pop)
plt.show()
```

### Scatter Plot

```python
plt.scatter(year, pop)
plt.show()
```

```python
x = np.random.randint(0,50,100)
y = np.random.randint(0,100,100)
```

```python
plt.scatter(x, y)
```

### Histogram

```python
l = np.random.randint(1,100, size = 100)
```

```python
plt.hist(l,bins = 5)
plt.show()
```

### Bar Chart

```python
plt.bar([1,3,5,7],[2,8,9,10], label = "bar chart example")
plt.bar([2,4,6,8],[3,5,11,12], label = "2nd example", color = 'g')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Bar chart')
plt.legend()
plt.show()
```

### Stack Plot

```python
days = [1,2,3,4,5]
sleeping = [7,6,7,6,9]
eating = [1,2,1,2,1]
working = [4,6,8,7,6]

# for adding labels
plt.plot([],[], color = 'm', label = 'sleeping', linewidth=5)
plt.plot([],[], color = 'c', label = 'Eating', linewidth=5)
plt.plot([],[], color = 'r', label = 'Working', linewidth=5)
plt.stackplot(days,sleeping, eating, working, colors = ['m','c','r'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Area Plot')
plt.legend()
plt.show()
```

### Pie Chart

```python
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']

plt.pie(slices,
       labels = activities,
       colors = cols,
       startangle=90,
       explode = (0,0.1,0,0.0025),
       autopct='%1.1f%%')
plt.title('Pie Plot')
plt.show()

```
